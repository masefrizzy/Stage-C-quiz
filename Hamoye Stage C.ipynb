```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
df = pd.read_csv("Data_for_UCI_named.csv")
df.head()
```

```python
# Drop stab field Because of the direct relationship between 'stab' and 'stabf' 
df.drop("stab", axis=1, inplace=True)
```

```python
df.head()
```

```python
df.stabf.value_counts()
```

```python
#Select the predictive and target columns

X = df.drop(columns="stabf")
y = df["stabf"]
```

```python
#split the data into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
y_train.value_counts()
```

```python
#encode categorical variable
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
# Use the standard scaler to transform the train set (x_train, y_train) and the test set (x_test).

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# x_train
normalised_xtrain_df = scaler.fit_transform(x_train)
normalised_xtrain_df = pd.DataFrame(normalised_xtrain_df, columns=x_train.columns)
# y_train
normalised_ytrain_df = scaler.fit_transform(y_train)
normalised_ytrain_df  = pd.DataFrame(normalised_ytrain_df, columns=y_train.columns)
# x_test
normalised_xtest_df = scaler.fit_transform(x_test)
normalised_xtest_df = pd.DataFrame(normalised_xtest_df, columns=x_test.columns)
```
